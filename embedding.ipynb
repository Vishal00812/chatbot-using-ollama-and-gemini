```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import FAISS
from langchain_community.document_loaders import PyPDFDirectoryLoader
```

```python
loader = PyPDFDirectoryLoader("./class12") 
```

```python
docs=loader.load()
```

```python
docs
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
final_documents = text_splitter.split_documents(docs)  
```

```python
from langchain_cohere import CohereEmbeddings
```

```python
import os
from dotenv import load_dotenv
```

```python
load_dotenv()
Cohere_api_key=os.getenv("COHERE_API_KEY")
```

```python
embeddings = CohereEmbeddings(model="multilingual-22-12")
```

```python
final_documents
```

```python
vectors = FAISS.from_documents(final_documents, embeddings)
```

```python
vectors
```

```python
index_file = "faiss_index_12"
```

```python
vectors.save_local(index_file)
```

```python
vectors = FAISS.load_local(index_file,CohereEmbeddings(model="multilingual-22-12"),allow_dangerous_deserialization=True)
```
